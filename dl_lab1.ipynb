```python
import numpy as np

# creating 1D array using numpy
a1 = np.array([1, 2, 3])
print(a1.shape)

#creating 2D array using numpy
a2 = np.array([[1, 2, 3],
               [4, 5, 6]])
print(a2.shape)

# creating 3D array using numpy
a3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(a3.shape)
```

```python
import torch

#creating 1D array using pytorch
t1 = torch.tensor([1, 2, 3])
print(t1.shape)

#creating 2D array using pytorch
t2 = torch.tensor([[1, 2, 3],
                   [4, 5, 6]])
print(t2.shape)

#creating 3D array using pytorch
t3 = torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(t3.shape)
```

```python
# operations using numpy
x = np.array([5, 6, 8])
y = np.array([9, 11, 12])

print("Addition of X and Y =", x + y)
print("substraction of X and Y =", x - y)
print("multiplication of X and Y =", x * y)
print("devision of X and Y =", x / y)

```

```python
# operations using pytorch
x = torch.tensor([5, 6, 8])
y = torch.tensor([9, 11, 12])

print("Addition of X and Y =", x + y)
print("substraction of X and Y =", x - y)
print("multiplication of X and Y =", x * y)
print("devision of X and Y =", x / y)

```

```python
# dot product using numpy
d= np.dot([5, 6, 8], [9, 11, 12])
print("dot product =",d)

#matrix multiplication using numpy
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print(" matrix multiplication ",A @ B)
```

```python
#  indexing using numpy
a = np.array([[10, 20, 30],
              [40, 50, 60]])

print(a[0, 1])      # 20
print(a[:, 1])      # column
print(a[1, :])      # row
```

```python
# indexing using pytorch
t = torch.tensor([[10, 20, 30],
                  [40, 50, 60]])

print(t[0, 1])
print(t[:, 1])
```

```python
#boolean masking using numpy
a = np.array([1, 2, 3, 4, 5])
a[a > 3]
```

```python
# boolean masking using pytorch
t = torch.tensor([1, 2, 3, 4, 5])
t[t > 3]
```

```python
# PyTorch: .view(), .reshape(), .unsqueeze(), .squeeze()
t = torch.arange(6)

print(t.view(2, 3))
print(t.reshape(3, 2))
print(t.unsqueeze(0))  # add dimension at index 0
print(t.squeeze())     # remove size-1 dimensions

```

```python
# in place vs out of place
x = torch.tensor([1, 2, 3])
y = x + 1
print(y)
```
